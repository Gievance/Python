```python
#np.tile 是 NumPy 库中的一个函数，它用于复制数组，沿指定的轴重复这些复制。它的语法如下：
#np.tile(A, reps)
#A: 是要重复的数组。
#reps: 定义了每个维度上重复的次数，可以是一个整数，也可以是一个元组。如果 reps 是一个整数，数组会在所有维度上重复相同次数。如果 reps 是一个元组，每个维度可以指定不同的重复次数。

import numpy as np

arr = np.array([[1, 2], [3, 4]])
tiled_arr = np.tile(arr, (2, 3))
print(tiled_arr)
```

```python
#np.sum 是 NumPy 库中用于计算数组元素和的函数。它可以计算数组中的所有元素之和，或者沿着指定的轴计算部分和。
#np.sum(a, axis=None)
#a: 要进行求和操作的数组。
#axis (可选参数): 指定沿着哪个轴计算和(0:列，1:行)。如果不提供该参数，函数将会对整个数组进行求和。
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

# 计算整个数组的和
total_sum = np.sum(arr)
print("总和:", total_sum)

# 沿着列的方向对数组进行求和
column_sum = np.sum(arr, axis=0)
print("列求和:", column_sum)

# 沿着行的方向对数组进行求和
row_sum = np.sum(arr, axis=1)
print("行求和:", row_sum)
```

```python
#np.argmin 是 NumPy 库中用于找出数组中最小元素索引的函数。它返回数组中最小值的索引，可以沿着指定的轴进行计算。
#np.argmin(a, axis=None)
#a: 要进行查找最小值索引的数组。
#axis (可选参数): 沿着指定轴查找最小值的索引(0:列，1：行)。如果不提供该参数，函数将会对整个数组进行查找。
import numpy as np

arr = np.array([[3, 7, 5],
                [8, 4, 2]])

# 查找整个数组中最小值的索引
min_index = np.argmin(arr)
print("整个数组最小值的索引:", min_index)

# 沿着列的方向查找最小值的索引
column_min_index = np.argmin(arr, axis=0)
print("列方向最小值的索引:", column_min_index)

# 沿着行的方向查找最小值的索引
row_min_index = np.argmin(arr, axis=1)
print("行方向最小值的索引:", row_min_index)
```

```python
#newCentroids = pd.DataFrame(dataSet).groupby(minDistIndices).mean()
#pd.DataFrame(dataSet): 这部分代码首先将 dataSet 转换为 Pandas 的 DataFrame。
#groupby() 方法根据指定的列 minDistIndices 中的值对数据进行分组。它创建了一个分组对象，用于对数据进行分组操作。
#mean在分组对象上调用了 mean() 方法。这意味着对每个分组中的数值数据进行求均值操作。
import pandas as pd
import numpy as np

# 创建一个示例的数据集
dataSet = {
    'Feature1': [1, 2, 4, 5, 8],
    'Feature2': [3, 1, 7, 6, 5],
    'Feature3': [10, 9, 12, 8, 15]
}

# 假设有一个表示分组的指标
minDistIndices = [0, 0, 1, 1, 0]
print(minDistIndices)
# 转换为 DataFrame
df = pd.DataFrame(dataSet)
print(df)
# 根据 minDistIndices 分组并计算平均值
newCentroids = df.groupby(minDistIndices).mean()

print(newCentroids)
```

```python
#导包
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#计算欧拉距离:返回[d1,d2,d3],分别为样本到三个簇的欧式距离
def calDist(dataset,center,k):
    clalist=[]#结果表，存储每个样本到k个簇的距离
    for data in dataset:
        dis=np.tile(data,(k,1))-center 
#         print(dis)
        dis=dis**2 #平方
        dis=np.sum(dis,axis=1)#按行求和
#         print(dis)
        dis==dis**0.5 #开平方
        clalist.append(dis)
    clalist=np.array(clalist)
    return clalist
# print(calDist([[1,2],[2,3]],[[1,2],[2,3]],2))

#计算质心
def calcen(dataset,center,k):
    distlist=calDist(dataset,center,k)
    mindistindex=np.argmin(distlist,axis=1)#最小距离索引=簇
    
    newcenter=pd.DataFrame(dataset).groupby(mindistindex).mean()#转换成DataFrame从而进行分组(默认按行分组)，然后取平均
    newcenter=newcenter.values
    changed=newcenter-center #计算质心的改变量
    return changed,newcenter
    
# changed,newcenter=calcen(createDataset(),random.sample(createDataset(),2),2)
# [changed,newcenter]
```

```python
#定义K-mean算法流程
def kmeans(dataset,k):
    #随机取k个样本作为初始质心
    center=random.sample(dataset,k)
    
    #更新一次质心，并将质心改变量作为循环终止条件
    
    changed,newcenter=calcen(dataset,center,k)
    while np.any(changed!=0):
        
        #定义簇群
        cluster=[]
        #计算样本到k个簇的距离
        distlist=calDist(dataset,newcenter,k)
        #按行，统计最小距离，相对应的列索引值=相对应簇
        mindistIndex=np.argmin(distlist,axis=1)
        
        #将样本插入簇
        for i in range(k):
            cluster.append([])
        for i,j in enumerate(mindistIndex):#i：索引值，j：簇索引
            cluster[j].append(dataset[i])
        
        #更新质心
        changed,newcenter=calcen(dataset,newcenter,k)
        
    #返回质心，簇群
    return newcenter,cluster
```

```python
#测试K-means算法
def createDataset():
    dataset=[
        [2,1],
        [1,2],
        [2,2],
        [3,2],
        [2,3],
        [3,3],
        [2,4],
        [3,5],
        [4,4],
        [5,3],
    ]
    return dataset

dataset=createDataSet()
newcenter,cluster=kmeans(dataset,2)
print(f"质心:{pd.DataFrame(newcenter)}")
print(f"集群：{cluster}" )
```
