多元线性回归

```python
import numpy as np
X=2*np.random.rand(100,1)
Y=3*X+4+np.random.randn(100,1)
```

```python
#制作散点图
import matplotlib.pyplot as plt

def plot_scatter(x,y):
    plt.scatter(x,y)
    plt.show()
    return 
plot_scatter(X,Y)
```

```python
#拟合
import numpy as np
X_b=np.c_[np.ones((100,1)),X] #加一列1
#print(X_b)
theta_best=np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)
theta_best
```

```python
#拟合直线和散点图放一起
Y2=theta_best[0]*X+theta_best[1]

def plot_scatter_line(x,y,y2):
    plt.scatter(x,y)
    plt.plot(x,y2,c='r')
    plt.show()
    return 
plot_scatter_line(X,Y,Y2)
```
