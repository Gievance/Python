利用numpy进行梯度下降


```python
#导包
import numpy as np
import matplotlib.pyplot as plt
```

求$y=x^{2}+2x+5$的最小值

```python
def plot_line():
    x=np.linspace(-6,4,100)#x轴范围
    y=x**2+2*x+5
    plt.plot(x,y)
#plt.show()
```

```python
#初始化 起始点x0,学习率theta和迭代次数
#第一次以固定值来演示
x=3
theta=0.02
iter=10
#y的导数
y_grad=2*x+2
```

```python
#y#y已经被赋值过了
```

```python
#计算y值的函数
def calculate_y(x):
    return x**2+2*x+5
```

```python
#梯度计算
def calculate_cur(x,y_grad,theta,iter):
    output_x=[]
    output_y=[]
    for i in range(iter):
        x=x-theta*y_grad
        output_x.append(round(x,2))
        output_y.append(calculate_y(round(x,2)))
    return output_x,output_y
output_x,output_y=calculate_cur(x,y_grad,theta,iter)
print(output_x,output_y)
for i in range(iter):
    plt.scatter(output_x,output_y)
plot_line()
plt.show()
```

演示可见，并没有到底


```python
#迭代至最优
#计算梯度
def calculate_grad(x):
    return 2*x+2
#计算最优
def calculate_cur(x,y_grad,theta):
    output_x=[]
    output_y=[]
    grads=calculate_grad(x)
    while grads>0.1:
        x=x-theta*grads
        output_x.append(round(x,2))
        output_y.append(calculate_y(round(x,2)))
        grads=calculate_grad(x)
    return output_x,output_y

output_x,output_y=calculate_cur(x,y_grad,theta)
#print(output_x,output_y)

def plot_scatter():
    plt.scatter(output_x,output_y)
    plot_line()
    plt.show()
    return 
plot_scatter()
```

如果学习率过高、过低会是什么现象？

```python
theta_high=0.4
def plot_scatter_theta(theta):
    output_x,output_y=calculate_cur(x,y_grad,theta)
    plt.scatter(output_x,output_y)
    plot_line()
    plt.show()
    
    return 
plot_scatter_theta(theta_high)
```
