# 访问外部Python 文件，并获得外部文件的局部变量

```python
import os
```

```python
version_file='outer/version.py' # 外部python文件

def get_version():
    with open(version_file,'r') as f:
        exec(compile(f.read(),version_file,'exec'))
    return locals()['__version__']

var=get_version()
print(var)
```

# 获取系统的所有环境变量及其对应的值

```python
env_variables =os.environ
for k,v in env_variables.items():
        print(f"{k}:{v}\n")
```
