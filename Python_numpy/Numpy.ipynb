# 1.2多维数据机构ndarray的基础
Numpy没有使用Python的列表，而是使用numpy自己实现的ndarray的数据结构

所以，转换数据类型也是要考虑的。

ndarray的官方定义：ndarray是由多个具有相同类型和尺寸的元素所组成的（通常具有固定的尺寸）多维的容器。

1.ndarray的元素必须是相同的类型、尺寸大小相同的数据

2.ndarray不像列表那样具有允许同时保存不同类型数据的灵活性，而且数组的大小也是不可变的。

```python
# ndarray属性
# 调用形式：(实例变量名).(属性)
a=np.array([1,2,3])
b=np.array([[1,2,3],[2,3,4]])
c=np.array([1.-2.6j,2.1+3j,4.-3.2j])
print(f"a:{a}")
print(f"b:{b}")
# 1、T：转置
print(f"a.T:{a.T}")
print(f"b.T:{b.T}")
# 2、data：数组中的数据从哪里开始的Python缓冲区对象，显示内存中的地址
print(f"a.data:{a.data}")
# 3、dtype：ndarray中元素的类型
print(f"a.dtype:{a.dtype}")


# 4、flags：ndarray中数据在内存中的保存方式的信息
print(f"a.flags:{a.flags}")
# 5、flat：将ndarray转换成一维数组的迭代器
print(f"b.flat[2]:{b.flat[2]}")


# 6、imag：ndarray中的虚数部分
print(f"c.imag:{c.imag}")
# 7、real：ndarray中的实数部分
print(f"c.real:{c.real}")


# 8、size：ndarray中所包含元素的数量
print(f"b.size:{b.size}")
# 9、itemsize：保存在内存中的每个元素所需的以字节为单位的内存容量
print(f"b.itemsize:{b.itemsize}")
# 10、nbytes：该ndarray中所有元素所占内存空间的总字节数
print(f"b.nbytes:{b.nbytes}")
#注：a.size*a.itemsize=a.nbytes


# 11、ndim：ndarray中所包含的维数
print(f"b.ndim:{b.ndim}")
# 12、shape:使用元组表示ndarray的形状
print(f"b.shape:{b.shape}")
#13、strides:使用元组表示的在各个维度方向上要移动到下一个接邻的元素所需移动的字节数
print(f"b.strides:{b.strides}")


# 14、ctypes：用于操作ctypes模块的迭代器
# 15、base：ndarray的基类对象(用于表示引用的是哪里的内存数据)
e=a[:2]
print(f"e.base is a.base:{e.base is a.base}")
print(f"e.base is a:{e.base is a}")
```

`ndarray的内存布局`：使用ndarray类生成的实例在内存中是以一维数组的形式进行存储
的。其中作为登记信息的-部分,用于描述数据的类型、数组的形状（shape）等以一维数组的形式保存的,用于指定读取元素数据方式的数据称为元数据。在这些元数据的后面是以数据形式保存的数组元素的值。

ndarray采用行主序方式。
      
      行主序：指数组以一行为单位存储于内存中

    a[0][0]  a[0][1]  a[0][2] ; a[1][0]  a[1][1]  a[1][2] ; a[2][0]  a[2][1]  a[2][2]

        列主序：指数组以一列为单位存储于内存中

    a[0][0]  a[1][0]  a[2][0] ; a[0][1]  a[1][1]  a[2][1] ; a[0][2]  a[1][2]  a[2][2]


```python
# ndarray内存布局
a=np.random.randn(100,100)
b=np.array(a,order='C') #行主序
c=np.array(a,order='F') #列主序
b.strides,c.strides
```

```python
# ndarray步长
a=np.ones((1000,))
b=np.ones((1000*100))[::100]
print(f"a.strides:{a.strides}")
print(f"b.strides:{b.strides}") 

#通过np.copy，避免使用view进行计算，创建copy进行计算效率更高
b_copy=np.copy(b)
print(f"b_copy.strides:{b_copy.strides}")
```

```python
# ndarray广播
a=np.array([1,2,3])
b=np.array([[1,1,1],[2,2,2]])
print(f"b+a:{b+a}")
print(f"a+1:{a+1}")
print(f"b+1:{b+1}")
```

## 1.8 副本与视图的区别
副本和视图是两个选项，用于指定现有数组和原有数组之间的关系。
* 副本：使用与原有数组不同的内存空间，但是数据内容是相同的
* 视图：与原有数组引用的是同一个内存地址。对视图数组中的元素所做的更改也会反映在原有数组的数据中。

```python
a=np.array([1,2,3])
d=a.view() #创建视图
# 副本不改变原有数组
c=a.copy()
c[0]=100
print(f"c:{c}")
print(f"a:{a}")

# 视图改变原有数组
d[0]=100
print(f"d:{d}")
print(f"a:{a}")


```

```python
#代入
# 针对变量中保存的数组的情况
a=np.array([1,2,3])
b=a
print(f"id(a)==id(b):{id(a)==id(b)}")

c=a[:]
print(f"id(a)==id(c):{id(a)==id(c)}")
c[1]=22
print(f"a:{a}")
# 即使使用切片，提取数组的一部分，得到的还是视图
      
      
#运算
# 对于四则运算符＋、—、＊、／及幂运算＊，结果相同
#如果要在不创建副本的前提下进行运算操作,还可以使用专门用于计算的函数。加法使用add函数’减法使用subtraCt函数’
a=np.array([1,2,3])
c=a
a=a+1 #a+1创建a的副本，再赋值给a

print(f"a:{a}")
print(f"c:{c}")

a=np.array([1,2,3])
c=a
a+=1 #直接进行加法运算，不创建副本
print(f"a:{a}")
print(f"c:{c}")
```

```python
# 数组的扁平化
# flatten:创建经过一维化处理后的副本，并返回副本的引用
# ravel:返回内存地址与原始数组相同的view对象
a=np.random.randn(2,3,9)
b=a.ravel()
c=a.flatten()
a[0,0,0]=129
print(f"{a[0,0,0],b[0],c[0]}")

#使用了特殊的索引值，通常都会生成副本
a=np.random.randint(10,size=100)
f=a[np.arange(0,10,2)]
print(np.may_share_memory(a,f)) #np.may_share_memory 用于确认内存是否是共享的
```

```python
# 副本和视图的分辨方法
# may_share_memory:可以对参数中所制定的两个数组所引用的内存空间是否是同一个地址信息判断。
# base属性：
a=np.array([1,2,3])
b=a[0:1]
print(b.base is a)
```
